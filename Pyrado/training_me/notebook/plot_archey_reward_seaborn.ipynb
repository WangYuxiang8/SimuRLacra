```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme(style="darkgrid")
rs = np.random.RandomState(50)
rs
```

```python
f, axes = plt.subplots(1,4,figsize=(16,4),sharex=True,sharey=True)
for ax, s in zip(axes.flat, np.linspace(0, 4, 10)):
    cmap = sns.cubehelix_palette(start=s, light=0.9, as_cmap=True)
    # Generate and plot a random bivariate dataset
    x, y = rs.normal(size=(2, 10))
    sns.kdeplot(
        x=x, y=y,
        cmap=cmap, fill=True,
        clip=(-5, 5), cut=10,
        thresh=0, levels=15,
        ax=ax,
    )
    ax.set_axis_off()

ax.set(xlim=(-3, 3), ylim=(-5, 5))
f.subplots_adjust(0, 0, 1, 1, .08, .08)
plt.savefig('filename.pdf')
```
